# Simple GUI Caculator.

```python
from tkinter import *
root =Tk()
#setting the width and height.
root.geometry("450x500")
expression=""
# defining function that will set expression and answer to user.
def setexpression(num):
    global expression
    expression=expression+str(num)
    value.set(expression)

#defining Function to calculate expression.
def calculator():
    try:
        global expression
        answer=str(eval(expression))
        value.set(answer)
    except:
        value.set("Enter correct expression")
        expression=""
#Function to clear expression.
def clear():
    global expression
    expression=""
    value.set(expression)
large_font=('verdana',15)
small_font=('verdana',10)

#declering variable to take input.
value=StringVar(value="   ")

Entry(root,textvariable=value,font=large_font).grid(row=0,column=0,columnspan=4,ipadx=80)
# for buttons
Button(root,text="+",fg="black",command=lambda:
    setexpression("+"),height=5,width=10).grid(row=1,column=0,pady=10)
Button(root, text="-", fg="black", command=lambda:
    setexpression("-"), height=5, width=10).grid(row=2, column=0, pady=10)
Button(root, text="X", fg="black", command=lambda: 
    setexpression("*"), height=5, width=10).grid(row=3, column=0,pady=10)
Button(root, text="/", fg="black", command=lambda: 
    setexpression("/"), height=5, width=10).grid(row=4, column=0,pady=10)
Button(root, text="1", fg="black", command=lambda: 
    setexpression("1"), height=5, width=10).grid(row=1, column=1,pady=10)
Button(root, text="2", fg="black", command=lambda: 
    setexpression("2"), height=5, width=10).grid(row=1, column=2,pady=10)
Button(root, text="3", fg="black", command=lambda: 
    setexpression("3"), height=5, width=10).grid(row=1, column=3,pady=10)
Button(root, text="4", fg="black", command=lambda: 
    setexpression("4"), height=5, width=10).grid(row=2, column=1,pady=10)
Button(root, text="5", fg="black", command=lambda: 
    setexpression("5"), height=5, width=10).grid(row=2, column=2)
Button(root, text="6", fg="black", command=lambda: 
    setexpression("6"), height=5, width=10).grid(row=2, column=3,pady=10)
Button(root, text="7", fg="black", command=lambda: 
    setexpression("7"), height=5, width=10).grid(row=3, column=1,pady=10)
Button(root, text="8", fg="black", command=lambda: 
    setexpression("8"), height=5, width=10).grid(row=3, column=2,pady=10)
Button(root, text="9", fg="black", command=lambda: 
    setexpression("9"), height=5, width=10).grid(row=3, column=3,pady=10)
Button(root, text="0", fg="black", command=lambda: 
    setexpression("0"), height=5, width=10).grid(row=4, column=2,pady=10)
Button(root, text=".", fg="black", command=lambda: 
    setexpression("."), height=5, width=10).grid(row=4, column=1,pady=10)
#show the calculated value by the widget "="
Button(root,text="=",fg="black",command=calculator,height=5,width=10).grid(row=4,column=3,pady=10)

#Entry clear  widget at the bottum.
Button(root, text= "Clear", fg="black", command=clear, height=10, width=20).grid(row=5, column=1, pady=10)

root.mainloop()
   



```
